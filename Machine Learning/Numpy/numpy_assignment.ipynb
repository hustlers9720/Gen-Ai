# Task 5: Statistical Operations (Core Focus)

```python
import numpy as np

marks = np.array([78, 85, 90, 66, 72, 88, 95, 60])  # Create a NumPy array containing marks of students

print("Mean", np.mean(marks))  # np.mean(marks) calculates the average value of all numbers in the array
print("Median", np.median(marks))  # np.median(marks) finds the middle value when the array elements are sorted
print("Variance", np.var(marks))  # np.var(marks) measures how far the numbers are spread out from the mean
print("Standard Deviation", np.std(marks))  # np.std(marks) calculates the square root of variance, showing the spread of data
print("Minimum and Maximum", np.min(marks), np.max(marks))  # np.min(marks) returns the smallest value in the array and # np.max(marks) returns the largest value in the array
print("Range(max-min)", np.max(marks) - np.min(marks))  # Range = Maximum value - Minimum value, showing the spread between highest and lowest marks
```

# Task 6: Percentiles & Sorting

```python
marks = np.array([78, 85, 90, 66, 72, 88, 95, 60])   # Array of student marks

# Sort the array
sorted_marks = np.sort(marks)  
print("Sorted Marks:", sorted_marks)  # np.sort() arranges the elements of the array in ascending order

# Calculate percentiles
p25 = np.percentile(marks, 25)  # Finds the value below which 25% of the data falls
p50 = np.percentile(marks, 50)  # Finds the 50th percentile (same as the median)
p75 = np.percentile(marks, 75)  # Finds the value below which 75% of the data falls

print("25th Percentile:", p25)
print("50th Percentile:", p50)
print("75th Percentile:", p75)

# Find average marks
average = np.mean(marks)  # Calculates the average (mean) of the marks

# Count students scoring above average
count_above_avg = np.sum(marks > average)  
# marks > average creates a boolean array (True/False)
# np.sum() counts how many True values (students above average)

print("Average Marks:", average)
print("Number of students scoring above average:", count_above_avg)
```

# Task 7: Mini Use Case – Sales Analysis

```python
# Given daily sales data
sales = np.array([1200, 1500, 900, 2000, 1800, 1700, 1600])  # NumPy array storing sales for 7 days

totalWeeklySales = np.sum(sales)  # np.sum() adds all elements to get total weekly sales
print("Total Weekly Sales:", totalWeeklySales)

averageDailySales = np.mean(sales)  # np.mean() calculates the average daily sales
print("Average Daily Sales:", averageDailySales)

highestSales = np.max(sales)  # np.max() finds the highest sales value
lowestSales = np.min(sales)  # np.min() finds the lowest sales value
print("Highest Sales:", highestSales, "Lowest Sales:", lowestSales)

standardDeviationofSales = np.std(sales)  # np.std() measures how much sales vary from the average
print("Standard Deviation of Sales:", standardDeviationofSales)

aboveAverageDays = []  # list to store sales values that are above the average
for i in sales:  # loop through each day's sales
    if i > averageDailySales:  # check if sales are greater than the average
        aboveAverageDays.append(i)  # add that sales value to the list

print("Sales Above Average:", aboveAverageDays)  # print all sales values greater than average
```
