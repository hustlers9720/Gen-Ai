# Task - 2 Constructor & Encapsulation

```python
# Modify the Product class to use a constructor.
# Make price a private attribute (__price).
# Create getter and setter methods:

    # get_price()

    # set_price(new_price) → should update only if new_price > 0

# Test modifying the price using the setter.

class Product:

    def __init__(self, name, price, category):
        self.name = name
        self.__price = price   # private attribute , here price is a private attribute and we define it via double underscore(__) before the name of it 
        self.category = category

    def get_info(self):
        print(f"Product Name: {self.name}, Price: {self.__price}, Category: {self.category}")

    def get_price(self):
        return self.__price # this is the getter function or method by which we get the value of private attribute

    def set_price(self, new_price):
        if new_price > 0:   # this is the setter function or method by which we set the value of the private attribute
            self.__price = new_price
        else:
            print("Price must be positive.")
```

```python
p1 = Product("Laptop", 1000, "Electronics")
p1.get_price()
```

```python
p1.set_price(1200)
p1.get_price()
```

# Task -3 : Inheritance (Single-Level)

```python
# Create a subclass ElectronicProduct that inherits from Product.
# Add an additional attribute: warranty_years.
# Override the get_info() method to include warranty information.
# Create an object and demonstrate inheritance and method overriding.

class Product:

    def __init__(self ,name , price , category):
            self.name = name
            self.price = price
            self.category = category

    def get_info(self):
          print(f"Product Name: {self.name} , Price: {self.price} , Category: {self.category}")


class ElectronicProduct(Product): # the child class that inherits the product parent class 

    def __init__(self, name, price, category , waranty_years):
            super().__init__(name, price, category)          
            self.waranty_years = waranty_years # it is the additional attribute that we add in the child class

    def get_info(self):
          print(f"Product Name: {self.name} , Price: {self.price} , Category: {self.category} , Waranty Years: {self.waranty_years}")
```

# Task-4 : Polymorphism

```python
# Create two classes:
# Laptop(Product)
# Mobile(Product)
# Override the get_info(self) method in both classes (print details in their own style).

class Product:

    def __init__(self ,name , price , category):
            self.name = name
            self.price = price
            self.category = category

    def get_info(self):
          print(f"Product Name: {self.name} , Price: {self.price} , Category: {self.category}")


class Laptop(Product):

      def __init__(self , name , price , category , waranty_years ,company):
            super().__init__(name, price, category)          
            self.waranty_years = waranty_years
            self.company = company

      def get_info(self):
          print(f"Product Name: {self.name} , Price: {self.price} , Category: {self.category} , Waranty Years: {self.waranty_years} , Company: {self.company}")      



class Mobile(Product):

      def __init__(self , name , price , category , waranty_years , company , camera_megapixels):
            super().__init__(name, price, category)          
            self.waranty_years = waranty_years
            self.company = company
            self.camera_megapixels = camera_megapixels

      def get_info(self):
          print(f"Product Name: {self.name} , Price: {self.price} , Category: {self.category} , Waranty Years: {self.waranty_years} , Company: {self.company} , Camera Megapixels: {self.camera_megapixels}")          
```

```python
dell = Laptop("Dell XPS 15" , 2000 , "Electronics" , 3 , "Dell")
dell.get_info()
```

```python
mi = Mobile("Mi 11" , 800 , "Electronics" , 2 , "Xiaomi" , 108)
mi.get_info()
```

```python
# Write a loop that iterates over objects of Laptop and Mobile and calls get_info() on each to demonstrate polymorphism.
products = [dell, mi]
for product in products:
    product.get_info()
```

# Task -5 Abstraction (Using Abstract Base Class)

```python
# Create an abstract class Payment with abstract method process_payment(amount).
# Create two subclasses:
# CreditCardPayment
# UPIPayment
# Override process_payment() in both subclasses using simple print statements.
# Test all classes.

from abc import ABC, abstractmethod

class Payment(ABC):

    @abstractmethod
    def process_payment(self, amount):
        pass


class CreditCardPayment(Payment):

    def process_payment(self, amount):
        print(f"Processing credit card payment of ${amount}.")

class UPIPayment(Payment):

    def process_payment(self, amount):
        print(f"Processing UPI payment of ${amount}.")        
```

```python
axisCreditCard = CreditCardPayment()
axisCreditCard.process_payment(100)
```

```python
phonePay = UPIPayment()
phonePay.process_payment(150)
```

# Task -6 : Magic Methods & Operator Overloading

```python
class Product:

    def __init__(self, name, price, category):
        self.name = name
        self.price = price
        self.category = category

    def get_info(self):
        print(f"Product Name: {self.name}, Price: {self.price}, Category: {self.category}")

    def __str__(self):
        return f"Product({self.name}, {self.price}, {self.category})"

    def __add__(self, other):
        if isinstance(other, Product):
            return self.price + other.price
        return NotImplemented
```

# Task -7 : Mini Project – Simple Inventory System (OOP Only)

```python
class Product:

    def __init__(self, name, price, category):
        self.name = name
        self.price = price
        self.category = category

    def __add__(self, other):
        return self.price + other.price

    def __str__(self):
        return f"Name: {self.name}, Price: {self.price}, Category: {self.category}"
```

```python
class Inventory:

    def __init__(self):
        self.products = []

    def add_product(self, product):
        self.products.append(product)

    def remove_product(self, name):
        for p in self.products:
            if p.name == name:
                self.products.remove(p)
                break

    def get_total_value(self):
        total = 0
        for p in self.products:
            total += p.price
        return total

    def show_all_products(self):
        for p in self.products:
            print(p)
```

```python
class Store:

    def __init__(self, store_name):
        self.store_name = store_name
        self.inventory = Inventory()

    def add_new_product(self, name, price, category):
        product = Product(name, price, category)
        self.inventory.add_product(product)

    def show_summary(self):
        print("Store Name:", self.store_name)
        print("Total Products:", len(self.inventory.products))
        print("Total Inventory Value:", self.inventory.get_total_value())
```

```python
store = Store("Tech Mart")

# 2. Adding 3 products
store.add_new_product("Laptop", 50000, "Electronics")
store.add_new_product("Mouse", 500, "Accessories")
store.add_new_product("Keyboard", 1500, "Accessories")
```

```python
# Show all products
store.inventory.show_all_products()
```

```python
# 3. Showing summary
store.show_summary()
```

```python
# 4. Using __add__ to combine prices of two products
p1 = store.inventory.products[0]
p2 = store.inventory.products[1]
```

```python
print("Combined Price of two products:", p1 + p2)
```
