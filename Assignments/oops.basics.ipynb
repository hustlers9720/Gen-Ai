```python
class Avenger :

    def __init__(self , name , show , catchpharse):
        self.name = name
        self.show = show
        self.catchpharse = catchpharse

    def introduction(self):
        print(f"Hello my name is {self.name}")

    def speak(self):
        print(f"{self.name} says {self.catchpharse}")
```

```python
Micky_Mouse = Avenger("Micky Mouse" , "Disney" , "Oh boy!")
Micky_Mouse.introduction()
Micky_Mouse.speak()
```

```python
class Avengerr:

    @staticmethod
    def mission():
        print("Save the earth") # this is the static method which doesnt need any self or cls keyword

    origin = "Avenger"
    @classmethod
    def show_origin(cls):
        print(f"Our origin is {cls.origin}")# this is the class method which need of the keyword "cls" to access the class variable

    def __init__(self, name, show, catchphrase):
        self.name = name
        self.show = show
        self.catchphrase = catchphrase

    def introduction(self):
        print(f"Hello my name is {self.name} and my show is {self.show} and my catchphrase is {self.catchphrase}")    
```

```python
captain = Avengerr("Captain America", "Marvel", "I can do this all day")
```

```python
captain.mission()
captain.show_origin()
captain.introduction()
```

```python
# inheritance
class Avenger :

    def __init__(self , name , power ):
        self.name = name
        self.power = power

    def introduction(self):
        print(f"Hello my name is {self.name}")

   

class Thor(Avenger):
    def __init__(self , name , power , weapon):
        super().__init__(name , power)
        self.weapon = weapon

    def weapon_details(self):
        print(f"My weapon is {self.weapon}")    
```

```python
thor = Thor("Thor", "Unlimited", "Hammer")
```

```python
thor.name
```

```python
thor.weapon
```

```python
thor.weapon_details()
```

```python
# Encapsulation 
class IronMan:
    def __init__(self , name , power , suit):
        self.name = name
        self.power = power
        self.suit = suit

    def introduction(self):
        print(f"My name is {self.name} and my power is {self.power} and my suit is {self.suit}")
```

```python
iron = IronMan("Adit" , "Genius" , "Mark 85")
```

```python
iron.introduction()
```

```python
# private members
class Hulk:
    def __init__(self , name, phone):
        self.name = name
        self.__phone = phone # this is a private member which cannot be accessed outside the class


     # getter method
    def get_phone(self):
        print(f"My phone number is {self.__phone}")  

    def set_phone(self , phone):
        self.__phone = phone
        print(f"My phone number is updated to {self.__phone}")     
    
```

```python
hulk = Hulk("Hulk" , "1234567890")
```

```python
hulk.get_phone()
```

```python
# absctract class and method
from abc import ABC, abstractmethod
class Vehicle(ABC):

    @abstractmethod
    def start_engine(self):
        pass

    @abstractmethod
    def stop_engine(self):
        pass

class Car(Vehicle):

    def start_engine(self):
        print("Car engine started")

    def stop_engine(self):
        print("Car engine stopped")
```
