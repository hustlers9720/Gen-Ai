# Task-1 :Basic Function: Price After Discount

```python
# 1. Write a function apply_discount(price, discount_percent) that:
# Returns the price after discount.
def apply_discount(price ,discount_percent):
    return price - (price* discount_percent/100)

apply_discount(2000,10)    
```

```python
# 2. If discount_percent is missing, apply a default discount of 5%.
def apply_discount(price ,discount_percent=5): # we have to give the default value if the discount percent is absent in the call of it.
    return price - (price* discount_percent/100)

apply_discount(1000)
```

```python
# 3.Test the function using:
# apply_discount(1000, 10)
# apply_discount(500) (uses default discount)

print(apply_discount(1000,10)) # it call the function with the function which having the both value
print(apply_discount(500)) # it call the function with the function having the default value of discount percent
```

```python
# 4. Extra (Optional):
# Add a condition to ensure the discount never exceeds 60%.
def apply_discount(price ,discount_percent=5):
    if discount_percent > 60: # this condition is to check that the discount percent should not be greater than 60
        discount_percent = 60 # if the discount percent is greater than 60 then it will set to 60
    return price - (price* discount_percent/100)
```

# Task-2 :Recursive Function: Factorial Utility

```python
# 1.Returns the factorial of n.
#2. Handles two edge cases:
def factorial(n):

    if(n<0):
        print( "Error: Factorial is not defined for negative numbers.")
        return None
    
    if(n==0):
        return 1
    
    return n*factorial(n-1)

factorial(6)
factorial(5)

factorial(0)
```

```python
# 3.Prints an error message if n is negative.
factorial(-3)
```

# Task 6 :Combined Utility Function

```python
# 1. Write a function process_prices(prices) that:
# function to process prices
def process_prices(prices):

    # apply 10% discount using map + lambda
    discounted_prices = list(map(lambda p: p - (0.10 * p), prices))

    # filter prices above 300
    filtered_prices = list(filter(lambda p: p > 300, discounted_prices))

    # return both lists
    return discounted_prices, filtered_prices


# test
d, f = process_prices([100, 500, 900, 50, 750])

print("Discounted Prices:", d)
print("Filtered Prices:", f)
```

# Task =7 : Mini Problem: Menu Using Functions

```python
# 1.Create the following three functions:

# add_price(prices_list, price) → adds a price to the list.
# get_average_price(prices_list) → returns the average price.
# get_max_price(prices_list) → returns the maximum price.
# Then create a simple menu:
# 1 → Add price
# 2 → Show average price
# 3 → Show highest price
# q → Quit
# Constraint:
# Use only loops + function calls (no OOP). 


    # function to add price
def add_price(prices_list, price):
    prices_list.append(price)

# function to get average price
def get_average_price(prices_list):
    if len(prices_list) == 0:
        return 0
    total = 0
    for p in prices_list:
        total += p
    return total / len(prices_list)

# function to get max price
def get_max_price(prices_list):
    if len(prices_list) == 0:
        return 0
    max_val = prices_list[0]
    for p in prices_list:
        if p > max_val:
            max_val = p
    return max_val


# menu program
prices = []

while True:
    print("\n1 - Add price")
    print("2 - Show average price")
    print("3 - Show highest price")
    print("q - Quit")

    choice = input("Enter choice: ")

    if choice == 'q':
        print("Program exited.")
        break

    elif choice == '1':
        val = input("Enter price: ")
        if not val.isdigit():
            print("Invalid input")
            continue
        add_price(prices, int(val))
        print("Price added.")

    elif choice == '2':
        print("Average Price:", get_average_price(prices))

    elif choice == '3':
        print("Highest Price:", get_max_price(prices))

    else:
        print("Invalid choice")
        continue
```
