# Task 1: Write Sales Records to a File

```python
# 1.Create a list of sales amounts:
sales = [1200, 450, 980, 1500, 3000]
# 2.Open a file named sales_data.txt in write mode.
# 3.Write each sale on a new line in the file.
f = open("sales_data.txt", "w") # by this operation we open the file if already exit , if not then it create and open for write operation
for s in sales:
    f.write(str(s) + "\n")
f.close() # by whis we have to close the file after the write operation is done.
```

```python
# 4.Close the file, then reopen it and print its contents.
f = open("sales_data.txt", "r")
print(f.read())
f.close()
```

```python
# optionally: Extra (Optional):
# .Write the data in comma-separated format instead of separate lines.
f = open("sales_data.txt", "w")
f.write(",".join(map(str, sales))) # by this we have to convert each element of the list to string and then write it with the commas in it.
f.close()
```

# Task 2 :Read File in Different Ways

```python
# 1.Using the same sales_data.txt:
with open("sales_data.txt", "r") as f: # this is the syntax as with by default it close the file , we have not to write the close() function. 
    data = f.read()
    print("Full File Content:")
    print(data)
```

```python
# 2.Read the first line using .readline().
with open("sales_data.txt", "r") as f:
    first_line = f.readline()
    print("\nFirst Line:")
    print(first_line)
```

```python
# 3.Read all lines using .readlines() and convert them into a list of integers.
with open("sales_data.txt", "r") as f:
    lines = f.readlines()

sales_list = []
for line in lines:
    nums = line.strip().split(",")   # split values
    sales_list.extend([int(x) for x in nums])

print("\nList of Sales Amounts:")
print(sales_list)
```

# Task 4 : Generate summary report from file

```python
# 1.Read all sales values from sales_data.txt.
# we know that "with" this we have to open the file as well as we dont need to close it , this (with) keyword take care of all the things.
with open("sales_data.txt" , "r") as f:
    content = f.read();

print(content)
```

```python
# 2.Convert them into integers.
with open("sales_data.txt" , "r") as f:
    content = f.read();

print(content) # with this i have all the content of the file 
sales_integer = [int(x) for x in content.strip().split(",")] # by this i have to convert the str into int to all value in the file.
print(type(sales_integer[0]))
```

```python
# 3.Calculate and print: Total Sales ,Highest Sale ,Lowest Sale ,Average Sale
total_sales = sum(sales_integer)
highest_sale = max(sales_integer)
lowest_sale = min(sales_integer)
average_sale = total_sales / len(sales_integer)

print(f"Total Sales: {total_sales}")
print(f"Highest Sale: {highest_sale}")
print(f"Lowest Sale: {lowest_sale}")
print(f"Average Sale: {average_sale}")
```

# Task 6 :Read File Safely (Error Handling Inside File Handling Only)

```python
# Ask the user for a filename to open.
# If the file exists, read and print it.
# If the file does not exist, print:
# "File not found. Please check the filename."
# Use simple condition checks with os.path.exists() only.

import os
user_filename = input("Enter the Filename to open")

if os.path.exists(user_filename):
    with open(user_filename, "r") as f:
        content = f.read()
    print(content)

else:
    print("File not found. Please check the filename.")
```

# Task 7 : Mini Project – Export Discounted Prices

```python
# create dictionary
prices = {
    "Mouse": 500,
    "Keyboard": 800,
    "Monitor": 7000,
    "Pendrive": 400,
    "Camera": 5000
}

# ask user for discount
discount = float(input("Enter discount percentage: "))

total_discounted = 0
count = 0

# write to file
with open("discount_report.txt", "w") as f:
    f.write("Product | Original Price | Discounted Price\n")

    for product, price in prices.items():
        discounted_price = price - (price * discount / 100)
        total_discounted += discounted_price
        count += 1

        f.write(f"{product} | {price} | {discounted_price}\n")

    # optional summary
    avg = total_discounted / count
    f.write(f"\nTotal Items: {count}\n")
    f.write(f"Average Discounted Price: {avg}\n")

# read file and print
print("\nReport Content:\n")
with open("discount_report.txt", "r") as f:
    print(f.read())
```
