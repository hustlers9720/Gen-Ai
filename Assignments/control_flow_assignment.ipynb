# Task - 1 : Discount Rules(id/elif/else)


```python
# 1.Write a Python program that:
#Reads an integer order_amount from the user using input().
order_amount = int(input("Enter the order amount:")) # this is the taken input from the user and cast it to integer
print(order_amount)
# Applies the following discount rules and prints the final amount:

if order_amount>=2000:# if order amount >=2000 then i have to applied 15 percent discount
    order_amount = order_amount-(0.15*order_amount)
elif order_amount>=1500 and order_amount<2000: # if order amount is in between then i have to applied 10 percent discount
    order_amount = order_amount-(0.10*order_amount)    
elif order_amount>=1000 and order_amount<1500: # if order amount is in between then i have to applied 5 percent discount
    order_amount = order_amount-(0.05*order_amount)
else:
    order_amount = order_amount # if order amount is less than 1000 then i have to applied no discount
print("The final order amount after discount is:",order_amount)   
```

```python
# Convert input to a number and handle invalid (non-numeric) input by displaying an error message and exiting the program.
try:
    order_amount = int(input("Enter the order amount:")) # this is the taken input from the user and cast it to integer
except ValueError: # if the user input is not a number then it will raise a value error and i have to handle it by displaying an error message and exiting the program
    print("Invalid input. Please enter a numeric value for the order amount.")
```

```python
# 2.Extra (Optional):
# Add a fixed 5% tax after discount.
# Print subtotal, tax, and final total.

subtotal = order_amount
tax = subtotal * 0.05
final_total = subtotal + tax

print("Subtotal:", subtotal)
print("Tax (5%):", tax)
print("Final Total:", final_total)
```

# Task -2 :Process Multiple Orders (for loop)

```python
# 1.Uses a for loop
# Applies the discount rules from Task 1 to each order
# Prints a summary table showing:
orders = [1200, 2500, 800, 1750, 3000]
for order in orders:
    if order>=2000:
        order_amount = order-(0.15*order)
    elif order>=1500 and order<2000:
        order_amount = order-(0.10*order)    
    elif order>=1000 and order<1500:
        order_amount = order-(0.05*order)
    else:
        order_amount = order
    subtotal = order_amount
    tax = subtotal * 0.05
    final_total = subtotal + tax
    print(f"Order Amount: {order}, Subtotal: {subtotal}, Tax (5%): {tax}, Final Total: {final_total}")
```

```python
# 2.Compute and print the total revenue after discounts 
total_revenue = sum(orders) - sum([order - (0.15*order if order >= 2000 else 0.10*order if order >= 1500 else 0.05*order if order >= 1000 else 0) for order in orders])
print("Total Revenue after discounts:", total_revenue)
```

```python
# 3.Extra (Optional):Print the number of orders that received a discount (discount > 0)
discount_count = 0
for order in orders:
    if order >= 1000:
        discount_count += 1
print("Number of orders that received a discount:", discount_count)
```

# Task 3 :User Menu (while loop + break/continue)

```python
# 1.Implement a simple while-loop driven menu that keeps asking the user to choose an action until they press q to quit.
# Menu options:
# 1 — Add order amount to a running list (use input())
# 2 — Show all orders and totals after applying discounts
# q — Quit

# create an empty list to store all order amounts
orders = []

# start an infinite while loop to keep showing menu
while True:

    # display menu options to user
    print("\nMenu:")
    print("1 - Add order amount")
    print("2 - Show orders and totals after discount")
    print("q - Quit")

    # take user choice as input
    choice = input("Enter your choice: ")

    # if user wants to quit
    if choice == 'q':
        print("Exiting program...")
        break   # break will stop the loop

    # option 1 -> add order amount
    elif choice == '1':
        amt = input("Enter order amount: ")

        # check if input is numeric
        if not amt.isdigit():
            print("Invalid amount! Please enter numbers only.")
            continue   # continue will re-show menu

        # convert to integer and add to list
        orders.append(int(amt))
        print("Order added successfully.")

    # option 2 -> show all orders and totals
    elif choice == '2':

        # check if list is empty
        if len(orders) == 0:
            print("No orders available.")
            continue

        total = 0

        # iterate through orders list
        for order in orders:

            # apply discount rules
            if order >= 2000:
                discount = 15
            elif order >= 1500:
                discount = 10
            elif order >= 1000:
                discount = 7
            else:
                discount = 0

            # calculate final amount after discount
            final_amount = order - (order * discount / 100)

            # add to total revenue
            total += final_amount

            # print summary for each order
            print(order, "->", str(discount) + "%", "->", final_amount)

        # print total revenue
        print("Total revenue after discount:", total)

    # if invalid menu choice entered
    else:
        print("Invalid choice! Try again.")
        continue
```

```python
# 2.Use:
# continue to re-show the menu after invalid input
# break to exit when the user enters q

# list to store order amounts
orders = []

# start while loop for menu
while True:

    # show menu
    print("\n1 - Add order amount")
    print("2 - Show orders and totals after discount")
    print("q - Quit")

    # take user input
    choice = input("Enter choice: ")

    # if user enters q -> exit program
    if choice == 'q':
        print("Program exited.")
        break   # break used to stop loop

    # option 1 -> add order
    if choice == '1':
        amt = input("Enter order amount: ")

        # check invalid input
        if not amt.isdigit():
            print("Invalid input! Enter numbers only.")
            continue   # continue used to re-show menu

        orders.append(int(amt))
        print("Order added.")

    # option 2 -> show orders
    elif choice == '2':

        # if no orders available
        if len(orders) == 0:
            print("No orders found.")
            continue   # go back to menu

        total = 0

        # loop through orders
        for order in orders:

            # apply discount rules
            if order >= 2000:
                discount = 15
            elif order >= 1500:
                discount = 10
            elif order >= 1000:
                discount = 7
            else:
                discount = 0

            # calculate final amount
            final = order - (order * discount / 100)
            total += final

            # print summary
            print(order, "->", str(discount) + "%", "->", final)

        print("Total revenue:", total)

    # invalid menu option
    else:
        print("Invalid choice! Try again.")
        continue   # continue to show menu again
```

# Task -4 : Loop Control with Conditions (break & continue)

```python
# 1. Given a list: daily = [200, 150, 0, 400, 50, -1, 300]
daily = [200, 150, 0, 400, 50, -1, 300]
total_sales = []
for amount in daily:
    if amount == -1:
        print("Corrupted Data")
        break
    elif amount ==0 :
        print("No sale day");
        continue
    else:
        total_sales.append(amount)
print("Total Sales:", sum(total_sales))
```
